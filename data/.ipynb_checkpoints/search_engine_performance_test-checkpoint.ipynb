## Normal search with 20 unique keywords (Sequentiel)

````python
# Number of tests
N = 200
# Keywords to search
keywords = ['general', 'man', 'book', 'john', 'west', 'company', 'york', 'illustrated', 'digest', 'pub', 
            'magazine', 'law', 'national', 'story', 'serie', 'post', 'publish', 'life', 'united', 'edited']
# Elasticsearch search request times (ms)
es_times = {}
# MySql indexation search request times (ms)
sql_times = {}
````

````python
print("Computing search times for Elasticsearch...")
def worker_es(k):
    """worker function"""
    tmp = [requests.get(f"{API_URL}/books/public?search={k}").json()['requestTime'] for i in range(N)]
    res = {}
    res['times'] = tmp
    res['key'] = k
    print(k)
    return res

#Assuming you want to use 3 processors
num_processors = 3#Create a pool of processors
p=multiprocessing.Pool(processes = num_processors)#get them to work in parallel
output = p.map(worker_es,keywords)

for o in output:
    es_times[o['key']] = o['times']
    Swagger
es_times
````

````python
print("Computing search times for MySQl index table...")
    
def worker_sql(k):
    """worker function"""
    tmp = [requests.get(f"{API_URL}/books/index?search={k}", headers = headers_protected).json()['requestTime'] for i in range(N)]
    print(k)
    res = {}
    res['times'] = tmp
    res['key'] = k
    return res

#Assuming you want to use 3 processors
num_processors = 3#Create a pool of processors
p=multiprocessing.Pool(processes = num_processors)#get them to work in parallel
output = p.map(worker_sql,keywords)

for o in output:
    sql_times[o['key']]=o['times']

sql_times
````

````python
f = open('MyEgrepTimes.json', "r")
egrep_times = json.loads(f.read())
f.close()
````

````python
es_means = {}
sql_means = {}
egrep_means = {}
egrep_stdev = {}
es_stdev = {}
sql_stdev = {}

for k in keywords:
    es_means[k] = mean(es_times[k])
    sql_means[k] = mean(sql_times[k])
    es_stdev[k] = stdev(es_times[k])
    sql_stdev[k] = stdev(sql_times[k])
    egrep_means[k] = mean(egrep_times[k])
    egrep_stdev[k] = stdev(egrep_times[k])
    
X_axis = np.arange(len(keywords))
````

````python
# build a bar plot and save it as a .png image
plt.figure(figsize=(15,10))
plt.bar(X_axis,[egrep_means[k] for k in keywords],0.4, yerr=[egrep_stdev[k] for k in keywords], capsize=5, label="MyEgrep")
plt.bar(X_axis + 0.2,[es_means[k] for k in keywords], 0.4, yerr=[es_stdev[k] for k in keywords], capsize=5, label="ES")
plt.bar(X_axis - 0.2,[sql_means[k] for k in keywords],0.4, yerr=[sql_stdev[k] for k in keywords], capsize=5, label="SQL")
plt.xlabel('Words')
plt.xticks(X_axis, keywords, rotation = 90)
plt.ylabel('Request time (ms)')
plt.title('Normal search with 20 unique keywords')
plt.yscale('log')
plt.legend()
plt.savefig('single_search.png', dpi = 300)
plt.show()
````

## Random words

````python
# Inializing random word list
word_site = "https://www.mit.edu/~ecprice/wordlist.10000"

response = requests.get(word_site)
WORDS = response.content.splitlines()
random_words = []
for i in range(100):
    random_words.append(WORDS.pop(random.randint(0, len(WORDS))).decode('ASCII'))
random_words[0]
````

````python
# Elasticsearch search request times (ms)
es_times3 = {}
# MySql indexation search request times (ms)
sql_times3 = {}
````

````python
requests.get(f"{API_URL}/books/index?search={random_words[0]}", headers = headers_protected).json()['requestTime']
````

````python
print("Computing search times for Elasticsearch...")
def rand_worker_es(k):
    """worker function"""
    tmp = [requests.get(f"{API_URL}/books/public?search={k}").json()['requestTime'] for i in range(10)]
    res = {}
    res['times'] = tmp
    res['key'] = k
    print(k)
    return res

#Assuming you want to use 3 processors
num_processors = 3#Create a pool of processors
p=multiprocessing.Pool(processes = num_processors)#get them to work in parallel
output = p.map(rand_worker_es,random_words)

for o in output:
    es_times3[o['key']] = o['times']
    
es_times3
````

````python
print("Computing search times for SQL...")
def rand_worker_sql(k):
    """worker function"""
    tmp = [requests.get(f"{API_URL}/books/index?search={k}", headers = headers_protected).json()['requestTime'] for i in range(10)]
    res = {}
    res['times'] = tmp
    res['key'] = k
    print(k)
    return res

#Assuming you want to use 3 processors
num_processors = 3#Create a pool of processors
p=multiprocessing.Pool(processes = num_processors)#get them to work in parallel
output = p.map(rand_worker_sql,random_words)

for o in output:
    sql_times3[o['key']] = o['times']
sql_times3
````

````python
es_means3 = {}
sql_means3 = {}
es_stdev3 = {}
sql_stdev3 = {}

for k in random_words:
    es_means3[k] = mean(es_times3[k])
    sql_means3[k] = mean(sql_times3[k])
    es_stdev3[k] = stdev(es_times3[k])
    sql_stdev3[k] = stdev(sql_times3[k])
    
X_axis3 = np.arange(len(random_words))
````

````python
# build a bar plot and save it as a .png image
plt.figure(figsize=(15,10))
plt.bar(X_axis3 + 0.2,[es_means3[k] for k in random_words], 0.4, yerr=[es_stdev3[k] for k in random_words], capsize=5, label="ES")
plt.bar(X_axis3 - 0.2,[sql_means3[k] for k in random_words],0.4, yerr=[sql_stdev3[k] for k in random_words], capsize=5, label="SQL")
plt.xlabel('Words')
plt.xticks(X_axis3, random_words, rotation = 90)
plt.ylabel('Request time (ms)')
plt.title('Normal search with random keywords')
plt.yscale('log')
plt.legend()
plt.savefig('random_simple_search_sql_equal.png', dpi = 300)
plt.show()
````
